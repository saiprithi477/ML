```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv(r'C:\Users\saiprithi\Documents\naresh it\datasience ai\xsel_sheets\winequality_red.csv')
x=df.drop('quality',axis=1)
y=df['quality']
df
```

**Lasso Regression**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lasso_reg=Lasso(alpha=0.1)
lasso_reg.fit(x_train,y_train)
```

```python
y_pred=lasso_reg.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print('mean square error:',mse)
```

**Ridge REgression**

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg=Ridge(alpha=0.1)
ridge_reg.fit(x_train,y_train)
```

```python
y_pred=ridge_reg.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print('mean square error:',mse)
```

**ElasticNet**

```python
from sklearn.linear_model import ElasticNet
ElasticNet()
```

```python
elastic_reg1=ElasticNet(alpha=0.1,l1_ratio=0.42)
elastic_reg1.fit(x_train,y_train)
```

```python
y_pred=elastic_reg1.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print('mean square error:',mse)
```
